```python
import pandas as pd
import numpy as np
```

```python
df_usd=pd.read_csv('https://gist.githubusercontent.com/attila5287/3e9ca129a6b29b0d91c4fc786434f1cb/raw/b84a2d1b65ee3d01f94f4a2fc0d6fcbb9fe00e91/currency.csv'
)
df_usd.head()
```

```python
df_usd.dtypes
```

```python
df_usd.drop(["INDICATOR", "SUBJECT", "MEASURE", "FREQUENCY", "Flag Codes"], axis=1, inplace=True)
df_usd
```

```python
df_usd[["Value"]]=df_usd[["Value"]].astype("float")
```

```python
df_usd.dtypes
```

```python
df_usd.rename(columns={"Value":"VALUE"}, inplace=True)
df_usd
```

```python
df_usd=df_usd.groupby(['TIME', 'LOCATION']).mean()
df_usd
```
